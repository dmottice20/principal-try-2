```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import os
```

### Read in the trasnformed data

```python
# Necessary variables for file exploration.
DATA_DIR = 'data'
TRANSFORMED_DIR = 'transformed'
feature_file = 'transformed_features.csv'
response_file = 'transformed_responses.csv'

# Read in the data.
feature_df = pd.read_csv('{}/{}/{}'.format(DATA_DIR, TRANSFORMED_DIR, feature_file))
response_df = pd.read_csv('{}/{}/{}'.format(DATA_DIR, TRANSFORMED_DIR, response_file))
feature_df.shape, response_df.shape
```

```python
# Convert date to str & set as index.
feature_df['DATE'] = pd.to_datetime(feature_df['DATE'])
feature_df.set_index('DATE', drop=True, inplace=True)
response_df['Date'] = pd.to_datetime(response_df['Date'])
response_df.set_index('Date', drop=True, inplace=True)
```

```python
# Plot the high level asset classes available.
macro_asset_classes = ['Equity Indices Market', 'Fixed Income Market', 'Commodities Market']
df = response_df[macro_asset_classes]
df.plot(subplots=False, figsize=(12,5))
plt.ylabel('Monthly Return (%)')
plt.savefig('returns.png')
plt.show()
```

```python
def grab_key(value, dictionary):
    """
    Return a value given the key of a dictionary.
    """
    found = False
    for i, v in enumerate(dictionary.values()):
        if v == value:
            found = True
            output = i
    
    if found:
        return output
    else:
        raise ValueError("{} is not found in {}".format(value, dictionary))
```
